```python
import pandas as pd
```

```python
def read():
    df = pd.read_excel("../../../data/MS/2023-4-13/23-5-17 Reciprocal/ms_list_of_test_applicants-20231220144255.xlsx")
    return df 
```

```python
df = read()
df
```

```python
df = df.rename(columns={"Unnamed: 1": "first_name", "Unnamed: 2": "last_name", "Unnamed: 4": "location"})
```

```python
df.location.unique()
```

```python
df.loc[:, "location"] = df.location.str.lower()

df = df[df.location.isin(["louisiana"])]

def clean_names(df):
    df.loc[:, "first_name"] = df.first_name.str.lower().str.strip()
    df.loc[:, "last_name"] = df.last_name.str.lower().str.strip()
    return df 

df = df.pipe(clean_names)
```
