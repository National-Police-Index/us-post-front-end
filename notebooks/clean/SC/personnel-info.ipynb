```python
import pandas as pd
import sys
# Get the directory of the current script
sys.path.append("../")

from lib.uid import gen_uid
from lib.clean import clean_names, clean_races, clean_sexes
```

```python
def read_personnel():
    dfa = pd.read_excel("../../../data/SC/2023-7-24/export_officer_info-2023-07-24-20231220144255.xlsx")

    return dfa
```

```python
df = read_personnel()

df
```

```python
df.columns
```

```python
df = df.rename(columns={"Gender": "sex", "EEOC": "race"})

df.loc[:, "race"] = df.race.str.lower().str.strip().str.replace(r"caucasian", "white", regex=True)

df.loc[:, "sex"] = df.sex.str.lower().str.strip()

df
```

```python
df = df.rename(columns={"Last Name": "last_name", "First Name": "first_name", "Middle Name": "middle_name", })

df = df.pipe(clean_names, ["first_name", "last_name", "middle_name"])

df
```

```python
df.to_csv("../../../data/SC/2023-7-24/export_officer_inf_clean.csv", index=False)
```
