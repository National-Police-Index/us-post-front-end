```python
import pandas as pd
```

```python
def read():
    df = pd.read_excel("../../../data/TX/2022-10-18/tx_copy_of_active_po_licensees_with_proficiency_and_service_records_10-7-2022-20231220144255.xlsx")
    return df 
```

```python
df = read()
```

```python
df = df.rename(columns={"FNAME": "first_name", "LNAME": "last_name", "MNAME": "middle_name"})

df.loc[:, "first_name"] = df.first_name.str.lower().str.strip()
df.loc[:, "middle_name"] = df.middle_name.str.lower().str.strip()
df.loc[:, "last_name"] = df.last_name.str.lower().str.strip()
```
