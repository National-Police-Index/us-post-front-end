```python
import pandas as pd
```

```python
def read():
    df = pd.read_excel("../../../data/TX/23-9-29 (Unique IDs)/psr-data-services-20231220144255.xlsx")
    return df 
```

```python
df = read()
```

```python
df = df.rename(columns={"FNAME": "first_name", "MNAME": "middle_name", "LNAME": "last_name", "RACE": "race", "GENDER": "gender"})
```

```python
def clean_dates(df):
    # Convert the 'ST_DATE' column to datetime
    df['start_date'] = pd.to_datetime(df['ST_DATE'])

    # Format the dates as 'Month Day, Year'
    df['start_date'] = df['start_date'].dt.strftime('%m/%d/%Y')

        # Convert the 'ST_DATE' column to datetime
    df['end_date'] = pd.to_datetime(df['END_DATE'])

    # Format the dates as 'Month Day, Year'
    df['end_date'] = df['end_date'].dt.strftime('%m/%d/%Y')
    return df.drop(columns=["ST_DATE", "END_DATE"])

df = df.pipe(clean_dates)
```

```python
def clean_names(df):
    df.loc[:, "first_name"] = df.first_name.str.lower().str.strip()
    df.loc[:, "middle_name"] = df.middle_name.str.lower().str.strip()
    df.loc[:, "last_name"] = df.last_name.str.lower().str.strip()
    return df 

df = df.pipe(clean_names)
```
