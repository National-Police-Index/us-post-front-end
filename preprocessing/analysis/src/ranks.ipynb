```python
import pandas as pd

# Load the dataset
df = pd.read_csv("../data/input/misconduct.csv")

# Drop the 'allegation_count' column if it exists
df = df.drop(columns=["allegation_count"], errors='ignore')  # errors='ignore' allows the script to run even if the column doesn't exist

# Calculate total number of unique agencies
total_agencies = df['agcy_name'].nunique()

# Assuming your DataFrame is named df
# Group by 'agcy_name' and sum the 'allegation_num' for each agency
agency_allegation_counts = df.groupby('agcy_name')['allegation_num'].sum().reset_index()

# Rename the column to 'allegation_count'
agency_allegation_counts.rename(columns={'allegation_num': 'allegation_count'}, inplace=True)

# Merge the aggregated counts back into the original DataFrame
df = pd.merge(df, agency_allegation_counts, on='agcy_name', how='left')

# Create a rank column based on 'allegation_count'
agency_allegation_counts['allegation_rank'] = agency_allegation_counts['allegation_count'].rank(method='dense', ascending=False).astype(int)

# Format the rank as "rank/total_agencies"
agency_allegation_counts['allegation_rank'] = agency_allegation_counts['allegation_rank'].apply(lambda x: f"{x}/{total_agencies}")

# Merge the rank back into the original DataFrame
df = pd.merge(df, agency_allegation_counts[['agcy_name', 'allegation_rank']], on='agcy_name', how='left')

# Process 'action' column for consistency
df.loc[:, "action"] = df.action.str.strip().str.replace(r"^ (\w+)", r"\1", regex=True).str.replace(r"(\w+) $", r"\1", regex=True)

# Filter for "Revoke Certification" actions
revoke_certification_df = df[df['action'] == "Revoke Certification"]

# Group by 'agcy_name' and count the occurrences of "Revoke Certification" for each agency
revoke_certification_counts = revoke_certification_df.groupby('agcy_name').size().reset_index(name='revoke_count')

# Sort and rank 'revoke_count'
revoke_certification_counts['revoke_rank'] = revoke_certification_counts['revoke_count'].rank(method='dense', ascending=False).astype(int)

# Format the rank as "rank/total_agencies"
revoke_certification_counts['revoke_rank'] = revoke_certification_counts['revoke_rank'].apply(lambda x: f"{x}/{total_agencies}")

# Merge the rank back into the original DataFrame
df = pd.merge(df, revoke_certification_counts[['agcy_name', 'revoke_rank']], on='agcy_name', how='left')

# Replace NaNs in 'revoke_rank' with a placeholder or alternative logic, as needed
# df['revoke_rank'] = df['revoke_rank'].fillna('N/A')  # Example placeholder

df
df.allegation_rank_x.unique()

df.agcy_name.nunique()

# # Save the modified DataFrame
# df.to_csv("../data/output/misconduct_new.csv", index=False)
```
